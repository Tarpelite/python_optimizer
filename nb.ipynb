```python
import numpy as np
from scipy.optimize import minimize, fmin_l_bfgs_b
```

```python
def penalty_I_function(n):
    alpha = 1e-5
    def func(x):
        res = 0.00000
        for i in range(n):
            f = np.sqrt(np.array(alpha)) * (x[i] - 1)
            res += f**2
        
        f = np.sum(x**2) - 1/4
        res += f**2
        return res
    return func
```

```python
def extended_rosenbrock_function(n):
    def func(x):
        res = 0.00000
        i = 1
        while 2*i <= n:
            f1 = 10 * (x[2*i] - x[i-1]**2)
            f2 = 1 - x[2*i-1]
            res += f1**2 + f2**2
            i += 1
        return res
    return func
```
